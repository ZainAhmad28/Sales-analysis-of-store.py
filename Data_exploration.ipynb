# Exploratory Data Analysis

# Gender

```python
ax = sns.countplot(x = "Gender", data=df)
```

```python
ax = sns.countplot(x = "Gender", data=df )

for bars in ax.containers :
    ax.bar_label(bars)
```

```python
 df.groupby(["Gender"] , as_index = False) [ "Amount"].sum().sort_values(by = "Amount", ascending = False)
```

```python
sales = df.groupby(["Gender"] , as_index = False) [ "Amount"].sum().sort_values(by = "Amount", ascending = False)

sns.barplot( x = "Gender" , y= "Amount" , data =sales)
```

# Age

```python
df.columns
```

```python
ax = sns.countplot(x = "Age Group", data=df )

```

```python
ax = sns.countplot(x = "Age Group", data=df , hue = "Gender" ) 

for bars in ax.containers :
    ax.bar_label(bars)
```

# state

```python
df.columns
```

```python
sales = df.groupby(["State"] , as_index = False) [ "Orders"].sum().sort_values(by = "Orders", ascending = False).head(10)
sns.set(rc = {"figure.figsize" : (20,5)})
sns.barplot( x = "State" , y= "Orders" , data =sales)
```

```python
sales = df.groupby(["State"] , as_index = False) [ "Amount"].sum().sort_values(by = "Amount", ascending = False).head(10)
sns.set(rc = {"figure.figsize" : (20,5)})
sns.barplot( x = "State" , y= "Amount" , data =sales)
```

# Marital Status

```python
ax = sns.countplot(x = "Marital_Status", data=df  ) 

```

```python
ax = sns.countplot(x = "Marital_Status", data=df  ) 

for bars in ax.containers :
    ax.bar_label(bars)
```
